## Bonus ï¼š
I compare the residual sugar between white wine and red wine because fermentation is one of the ways to make wine and using residual sugar we can divide wines into dry wines or other kinds. I order the level from large to small, and found that red wine are lower than white wine in general. That's why most of time red wine will tastes more bitter than white wine.

And we can see nearly 1150 white wines are higher than 10 and only 10 red wines are higher than 10. So we can calculate about 23.48% in white wine are dry wine and only 0.6% of red wine is dry wine.

```python
import seaborn as sns
import numpy as np
from numpy.random import randn
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import stats

 
x = df_white.iloc[:,3] #the residual sugar value
y = df_red.iloc[:,3] #the residual sugar value

x = sorted(x,reverse=True)
y = sorted(y,reverse=True)
#print(len(y))

plt.figure(figsize=(10,6))
plt.plot(x[0:1599],label = 'white wine')
plt.plot(y[0:1599],label = 'red wine')
plt.xlabel('Observations')
plt.ylabel('Residual Sugar')
plt.grid(linestyle='-.')
plt.legend(fontsize='16')
plt.title("Top 1600 Observatoins")

```
