```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
res = requests.get("https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_WEEKLY&symbol=BTC&market=SGD&apikey= E50N4ZOEJM5HJXXZ")
res
```

```python
res_json = res.json()
#res_json
```

```python
data = pd.DataFrame.from_dict(res_json['Time Series (Digital Currency Weekly)']).T
data = data.iloc[0:50,:]
data = data.iloc[::-1]
data
```

```python
data = data.astype('float32')
```

```python
fig = plt.figure(figsize=(15,6)) 

plt.xticks(rotation=90) 
plt.plot(data['1a. open (SGD)'])
plt.title("Bitcoin rate Ranges in One Year(SGD)")
plt.grid()
```
